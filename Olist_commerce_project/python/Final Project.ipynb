## **Data Cleaning: Handling Missing Review Comments**

## **Data Transformation: Delivery Time Analysis**

```python
import numpy as np
orders = pd.read_csv("C:/Users/Dell/OneDrive/Desktop/Projects/Final Project/archive/olist_orders_dataset.csv")

orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])

orders['estimated_delivery_variance'] = (orders['order_delivered_customer_date'] - 
                                         orders['order_estimated_delivery_date']).dt.days


orders['delivery_flag'] = np.where(orders['estimated_delivery_variance'] <= 0, 'On Time',
                             np.where(orders['estimated_delivery_variance'] <= 5, 'Slightly Delayed', 
                                      'Severely Delayed'))

orders.to_csv("orders_enriched.csv", index=False)

print("Orders dataset transformed and saved successfully!")
```

## **Data Cleaning: Resolving Duplicate City Names in the Geolocation Dataset**

```python

df = pd.read_csv("C:/Users/Dell/OneDrive/Desktop/Projects/Final Project/archive/olist_geolocation_dataset.csv")

cleaned_df = df.groupby('geolocation_zip_code_prefix')['geolocation_city'] \
               .agg(pd.Series.mode) \
               .reset_index()

cleaned_df.to_csv("cleaned_geolocation.csv", index=False)

print("Duplicates resolved and cleaned file saved successfully!")
```
